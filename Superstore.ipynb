## Segment Analysis

```python
# Segment summary

segent_summary=df.groupby('Segment')['Sales','Profit','Quantity','Shipping Cost','DiscountAmount'].sum()

# calculate total sales and profit
total_sales=segent_summary['Sales'].sum().round(2)
total_profit=segent_summary['Profit'].sum().round(2)

# Add new columns to dataframe
segent_summary['% of Sales'] = segent_summary['Sales'].apply(lambda x: (x/total_sales)*100) 
segent_summary['% of Profit'] = segent_summary['Profit'].apply(lambda x: (x/total_profit)*100)
segent_summary['Profit Margin(%)']=segent_summary['Profit']/segent_summary['Sales']*100

#print the dataframe with 2 decimal places
segent_summary.round(2)
```

```python
plt.bar(segent_summary.index, segent_summary["Sales"])
plt.title("Total sales per product segment")
#plt.xlabel("Segment")
#plt.ylabel("Sales")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
# bar chart for profit
segent_summary.sort_values(by='Profit',ascending=True)
#plt.figure(figsize=(6,6))
plt.bar(segent_summary.index, segent_summary["Profit"])
plt.title("Total profit per product segment")
#plt.xlabel("Segment")
#plt.ylabel("Profit")
plt.show()
```

* Consumer segment contributed the most to both sales(51.48%) and profit(51.06%) with home office contributing the least in sales and profit.
* However home office has the highest profit margin(11.99%).

## Market Analysis

```python
#Summary of market performance 
market_summary=df.groupby('Market')['Sales','Profit','Quantity','Shipping Cost','DiscountAmount'].sum().round(2)

market_summary['% of Sales'] = market_summary['Sales'].apply(lambda x: (x/total_sales)*100)
market_summary['% of Profit'] = market_summary['Profit'].apply(lambda x: (x/total_profit)*100)
market_summary['Profit Margin(%)']=market_summary['Profit']/market_summary['Sales']*100

market_summary.round(2)
```

```python
# find the market with minimum profit margin

min_market=market_summary.loc[:,'Profit Margin(%)'].idxmin()
min_market_value=market_summary.loc[:,'Profit Margin(%)'].min()

print(f'Market with lowest profit margin is  {min_market} with {min_market_value:.2f}%')
```

```python
# find the market with maximum profit margin

max_market=market_summary.loc[:,'Profit Margin(%)'].idxmax()
max_market_value=market_summary.loc[:,'Profit Margin(%)'].max()

print(f'Market with highest profit margin is  {max_market} with {max_market_value:.2f}%')
```

```python
#Plot a bar chart for market sales and profit
plt.figure(figsize=(10,6))
market_summary.sort_values(by='Sales',ascending=False).plot.bar(y=["Sales","Profit"])
plt.title("Total Profit and Sales per product market")
plt.xlabel("Market")
plt.ylabel("Profit")
plt.ticklabel_format(style='plain', axis='y')
plt.legend(loc='upper right')
plt.show()
```

* APAC market has the highest sales(28.36%) and profit(29.71%) followed by the EU market
* Canada market has the lowerest sales(0.53%) and profit(1.21%). It however has the highest profit margin contribution(26.62%) compared to 12.16% for the APAC market.
* The low sales in Canada can be attributed to the low quantity of products sold. Efforts should by directed towards increasing company's visibility and sales in this market.
